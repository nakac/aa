# Read Excel

```python
data = pd.read_excel('Poly.xlsx')
X=np.array(data['x']).reshape(-1,1)
y=np.array(data['y']).reshape(-1,1)
```

# Plot data

```python
plt.plot(X, y, "b.")
```

# Import PolynomialFeatures

```python
from sklearn.preprocessing import PolynomialFeatures
```

# Degree 2

```python
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

# Check the 1st Row value in Excel 

```python
pd.DataFrame(X, columns=['X']).head(5)
```

```python
pd.DataFrame(X_poly, columns=['X', 'X^2']).head(5)
#pd.DataFrame(X_poly).head(5)
```

# Polynomial regression
* Fit the nonliner data to the linear model 

```python
model = LinearRegression()
model.fit(X_poly, y)
```

# Coefficient

```python
model.intercept_
```

```python
model.coef_
```

# Plot

```python
X_new = np.linspace(-3, 3, 100).reshape(100, 1)
X_new_poly = poly_features.transform(X_new)
y_new_regul = model.predict(X_new_poly)

plt.plot(X, y, "b.", ms=10)
plt.plot(X_new, y_new_regul,'g-', linewidth=3)
```
