# The Normal Equation
# Find wight $\theta_{0} ,\theta_{1}$ in  $ y=\theta_{0} +\theta_{1}x $

$ \Large
    X = \left(
    \begin{array}{ccc}
      1 & x_{1} \\
      \vdots &  \vdots \\
      1 & x_{100}
    \end{array}
  \right)
$

```python
X_b = np.c_[np.ones((100, 1)), X]  # add x0 = 1 to each instance
```

$\huge \theta = (X^{T} X)^{-1} (X^{T}  y)$

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```
